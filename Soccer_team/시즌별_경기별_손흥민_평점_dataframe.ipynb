### **15/16시즌 경기 중 프리미어 리그만 뽑아오기**

```python
from bs4 import BeautifulSoup
import urllib.request
import urllib.parse

web_url = 'https://www.skysports.com/tottenham-hotspur-results/2015-16'

epl_html = []


# BeautifulSoup으로 페이지 접속 후, div 태그에 있는 class 중 fixres__body의 내용을 찾는다
# epl_html에 추가
# soup.find_all('태그명', '클래스명')

with urllib.request.urlopen(web_url) as response:
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')    

    a = soup.find_all('div', 'fixres__body')

    epl_html.append(a)

    
# 이중 리스트 제거
epl_html = sum(epl_html, [])

# 리스트를 문자열로
epl_html = " ".join(map(str, epl_html))
```

```python
epl_html
```

### **15/16시즌 'epl_links' 정규화**

```python
import re

# "Premier League.*\s.*\s.*\s" --> 프리미어 리그로 시작하면서 \n(엔터)를 세번까지 찾고, epl_links에 저장
epl_links = re.findall("Premier League.*\s.*\s.*\s", epl_html)


# epl_links[i].replace('"', "-") --> 큰따옴표를 하이픈으로 대체
# epl_links[i][157:-3] --> 찾는 url 앞뒤의 값 제거 
for i in range(0, len(epl_links)):
    epl_links[i] = epl_links[i].replace('"', "-")
    epl_links[i] = epl_links[i][157:-3]
    
epl_links
```


### **15/16시즌 'date' 정규화**

```python
date_html = []

for i in epl_links:
  with urllib.request.urlopen(i) as response:
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')    

    a = soup.find_all('li', 'match-header__detail-item')

    date_html.append(a)

    
# 이중 리스트 제거
date_html = sum(date_html, [])

# 리스트를 문자열로
date_html = " ".join(map(str, date_html))
```

```python
#  pm으로 시작하고, 첫번째 엔터(\n)까지 찾는다
date = re.findall("pm.*\s", date_html)

# 'pm "삭제하고, 처음 등장하는 "<" 인덱스 까지 자른다
for i in range(0, len(date)):
  date[i] = date[i][3:]
  date[i] = date[i][:date[i].find("<")]
```

### **15/16시즌 '평점' 정규화**

```python
# epl_links 하나씩 넣어주면서 div class의 article tag 긁어오기

rating_html = []

for i in epl_links:
  with urllib.request.urlopen(i) as response:
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')    

    a = soup.find_all('div', 'article')

    rating_html.append(a)

# 이중 리스트 제거    
rating_html = sum(rating_html, [])

# 리스트를 문자열로
rating_html = " ".join(map(str, rating_html))
```

```python
# 큰따옴표를 하이픈으로 대체
rating_html = rating_html.replace('"', "-")

# rating_html에 정규표현식을 사용하여 Tottenham 선수들의 평점만 rating(list) 변수에 저장
rating = re.findall('<strong>Tottenham.*?</p>', rating_html)
```

```python
# rating list를 문자열로(정규표현식 사용위해)
#rating = " ".join(map(str, rating))

# 정규표현식으로 Son만 찾아낼 수 있으나 Son이 참가한 경기의 날짜를 알 수 없음...ㅠ
#re.findall("Son", rating)
```

```python
# 선수들 평점만 남기고 필요없는 문자열 제거
for i in range(0, len(rating)):
  rating[i] = rating[i].replace("\xa0", "")
  rating[i] = rating[i].replace(":", "")
  rating[i] = rating[i].replace("</p>", "")
  rating[i] = rating[i].replace(" ", "")
  rating[i] = rating[i].replace("<strong>Tottenham</strong>", "")
  rating[i] = rating[i].replace(";", ",")
  rating[i] = rating[i].replace(".", "")
  rating[i] = rating[i].replace("<strong>Tottenham", "")
  rating[i] = rating[i].replace("</strong>", "")
  rating[i] = rating[i].replace("Hotspur", "")
  print(rating[i])
```

### **18/19시즌 경기 중 프리미어 리그만 뽑아오기**

```python
from bs4 import BeautifulSoup
import urllib.request
import urllib.parse

web_url_1819 = 'https://www.skysports.com/tottenham-hotspur-results/2018-19'

epl_html_1819 = []


# BeautifulSoup으로 페이지 접속 후, div 태그에 있는 class 중 fixres__body의 내용을 찾는다
# epl_html에 추가
# soup.find_all('태그명', '클래스명')

with urllib.request.urlopen(web_url_1819) as response:
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')    

    a = soup.find_all('div', 'fixres__body')

    epl_html_1819.append(a)

    
# 이중 리스트 제거
epl_html_1819 = sum(epl_html_1819, [])

# 리스트를 문자열로
epl_html_1819 = " ".join(map(str, epl_html_1819))
```

```python
epl_html_1819
```

### **18/19시즌 'epl_links' 정규화**

```python
import re

# "Premier League.*\s.*\s.*\s" --> 프리미어 리그로 시작하면서 \n(엔터)를 세번까지 찾고, epl_links에 저장
epl_links_1819 = re.findall("Premier League.*\s.*\s.*\s", epl_html_1819)


# epl_links[i].replace('"', "-") --> 큰따옴표를 하이픈으로 대체
# epl_links[i][157:-3] --> 찾는 url 앞뒤의 값 제거 
for i in range(0, len(epl_links_1819)):
    epl_links_1819[i] = epl_links_1819[i].replace('"', "-")
    epl_links_1819[i] = epl_links_1819[i][157:-3]
    
epl_links_1819
```

```python
len(epl_links_1819)
```

### **18/19시즌 'date' 정규화**

```python
date_html_1819 = []

for i in epl_links_1819:
  with urllib.request.urlopen(i) as response:
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')    

    a = soup.find_all('li', 'match-header__detail-item')

    date_html_1819.append(a)

    
# 이중 리스트 제거
date_html_1819 = sum(date_html_1819, [])

# 리스트를 문자열로
date_html_1819 = " ".join(map(str, date_html_1819))

date_html_1819
```

```python
#  pm으로 시작하고, 첫번째 엔터(\n)까지 찾는다
date_1819 = re.findall("pm.*\s", date_html_1819)

# 'pm "삭제하고, 처음 등장하는 "<" 인덱스 까지 자른다
for i in range(0, len(date_1819)):
  date_1819[i] = date_1819[i][3:]
  date_1819[i] = date_1819[i][:date_1819[i].find("<")]
  
date_1819
```

### **18/19시즌 '평점' 정규화**

```python
# epl_links_1819 하나씩 넣어주면서 div class의 widge-fact__body tag 긁어오기
# widge-fact__body로 변경

rating_html_1819 = []

for i in epl_links_1819:
  with urllib.request.urlopen(i) as response:
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')    

    a = soup.find_all('div', 'widge-fact__body')

    rating_html_1819.append(a)

# 이중 리스트 제거    
rating_html_1819 = sum(rating_html_1819, [])

# 리스트를 문자열로
rating_html_1819 = " ".join(map(str, rating_html_1819))

rating_html_1819
```

```python
# rating_html_1819에 정규표현식을 사용하여 Tottenham 선수들의 평점만 rating_1819(list) 변수에 저장
# re.findall('Tottenham:.*?<br|Spurs:.*?<br', rating_html_1819) --> Tottenham:으로 시작해서 <br으로 끝나거나 Spurs:으로 시작해서 <br으로 끝나는 값 리턴
rating_1819 = re.findall('Tottenham:.*?<br|Spurs:.*?<br', rating_html_1819)
```
